```python
from urllib.parse import urlparse

def extract_website_name(link):
    parsed_link = urlparse(link)
    return parsed_link.netloc
arr1 = []
for i in range(len(df)):
  link=df['URL'][i]
  website_name = extract_website_name(link)
  arr1.append(website_name)
arr=np.array(arr1)
np.unique(arr)
```

```python


def Economictimes(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            article_div = soup.find('div', {'class': 'artText'})
            if article_div:
                article_content = article_div.text.strip()
                return article_content
    except Exception as e:
        print("Error:", e)
    return ""
```

```python

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}
def BusinessToday(url):
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        main_content = soup.find('div', class_='story-with-main-sec')
        if main_content:
            return main_content.get_text().strip()
        else:
            return ""
    else:
        return ""
```

```python
def Financial(url):
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        paragraphs = soup.find_all('p')
        texti=""
        for paragraph in paragraphs:
            texti+=paragraph.get_text()
        return texti    
    else:
        return ""
```

```python
def MoneyControl(url):
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        main_content = soup.find('div', class_='content_wrapper')
        texi=""
        if main_content:
            paragraphs = main_content.find_all('p')

            for paragraph in paragraphs:
                texi+=paragraph.get_text()
        else:
            return texi
        return texi
    else:
        return ""
```

```python
def Livemint(url):
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        main_content = soup.find('div', class_='mainArea')
        texi=""
        if main_content:
            paragraphs = main_content.find_all('p')
            for paragraph in paragraphs:
                texi+=paragraph.get_text().strip()
        else:
            return ""
        
        return texi

    else:
        return ""
```
